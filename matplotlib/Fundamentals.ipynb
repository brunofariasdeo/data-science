```python
import numpy as np
x = np.arange(0,10,0.1)
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15,10))
fig1.suptitle("Minha primeira figura", size=30)

f1_axes[0,0].plot(np.sin(x))
f1_axes[0,0].set_title("Caixa 1")
f1_axes[0,0].set_xlabel("Caixa 1 - X")
f1_axes[0,0].set_ylabel("Caixa 1 - Y")

f1_axes[0,1].plot(np.cos(x))
f1_axes[0,1].set_title("Caixa 2")

f1_axes[1,0].plot(np.sin(x), label="sin(x)", color="purple")
f1_axes[1,0].plot(np.cos(x), label="cos(x)", color="darkturquoise")
f1_axes[1,0].set_title("Caixa 3")
f1_axes[1,0].legend()

f1_axes[1,1].plot(np.tan(x))
f1_axes[1,1].set_title("Caixa 4")
```
